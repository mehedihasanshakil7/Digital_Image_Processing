# Import necessary modules

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Takegrayscale image and perform the following operations

```python
img = cv2.imread("kidney.tif", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(6, 4))
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Perform the brightness enhancement of a specific range of gray levels and observe its result

```python
start = 150
end = 240

enhanced_img1 = img.copy() # Out of range pixels will be black (0)
enhanced_img2 = img.copy() # Out of range pixels will be left unaltered

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if start <= img[i][j] <= end:
            enhanced_img1[i][j] = 255
            enhanced_img2[i][j] = 255
        else:
            enhanced_img1[i][j] = 0

plt.figure(figsize=(8, 8))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(enhanced_img1, cmap='gray')
plt.title('Brightness Enhanced Image\n(Out of range pixels Black)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(enhanced_img2, cmap='gray')
plt.title('Brightness Enhanced Image\n(Out of range pixels Unaltered)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## (b) Differentiate the results of power law transform and inverse logarithmic transform
## Power Law Transform, $s = c * r^y$, where $c = 1$
## Inverse Logarithmic Transform, $s = (e^{r/c} - 1)$, where $c = 255/log(1+255)$

```python
img = cv2.imread("runway.tif", cv2.IMREAD_GRAYSCALE)

power_law_img = np.power(img / 255.0, 4.0) * 255.0

c = 255 / np.log(1 + 255)
inverse_log_img = np.exp(img / c) - 1

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(power_law_img, cmap='gray')
plt.title('Power Law Transformed Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(inverse_log_img, cmap='gray')
plt.title('Inverse Logarithmic Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## (c) Find the difference image between original and the image obtained by last three bits (MSB)

```python
img = cv2.imread("dollar.tif", cv2.IMREAD_GRAYSCALE)

last_three_bits = img & 0b11100000

plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)

plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 1, 2)
plt.imshow(last_three_bits, cmap='gray')
plt.title('Image with Last Three Bits (MSB)')
plt.axis('off')

plt.subplot(3, 1, 3)
plt.imshow(img - last_three_bits, cmap='gray')
plt.title('Difference Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
